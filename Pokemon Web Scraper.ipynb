```python

# extracts the data from the pokemon url
#we will then fetch the desired answers using data frame functions. 

pokemon_url = "https://pokemondb.net/pokedex/all"
pokepage = requests.get(pokemon_url)


pokepage_soup = BeautifulSoup(pokepage.text, 'html.parser')


pokepage_list = pokepage_soup.find_all(name="table", class_="data-table block-wide")
pokemon_records = pokepage_soup.find_all("tr")

data = []


for row in pokemon_records:
    
    cols = row.find_all('td')
    cols = [ele.text for ele in cols]
    data.append([ele for ele in cols if ele]) 
```

```python
def row_to_df(row_number):
    
    pokemon =  pokemon_records[row_number].find_all(name = 'td', class_ = 'cell-num')
    
    pokemon_name = pokemon_records[row_number].find('a').text
    
    
    url = pokemon_records[row_number].find_all('a')    
    
    types = ''
    types=types+url[1].text
    if len(url) > 2:
        types=types+' '+url[2].text
    
    total_points = pokemon_records[row_number].find(name = 'td' , class_ = 'cell-total').text
    
    
    
    
    pokelist = []
    spcl_powers = []
    pokelist.append(pokemon_name)
    
    
    for i in pokemon:
        pokelist.append(i.text)
    final_url='https://pokemondb.net'+url[0]['href']
    pokelist.append(types)
    pokelist.append(total_points)
    pokelist.append(final_url)
    poke_dictionary_data = pd.DataFrame(pokelist)
    
    return poke_dictionary_data
    
```

```python
import numpy as np
bulbasaur=pd.DataFrame([])
for i in range(1,927):
    a_list=row_to_df(i).T
    bulbasaur=bulbasaur.append(a_list)
    
bulbasaur.index = np.arange(1, len(bulbasaur) + 1)        
bulbasaur
```
